Import delle librerie necessarie per la simulazione. 

```python
import numpy as np
import matplotlib.pyplot as plt
```

Simuliamo il segnale proveniente dallo sfondo come Gaussiano a media nulla. Simuliamo l'acquisizione su due canali (fase e quadratura) che rappresenteranno la parte reale e la parte immaginaria del segnale acquisito. Visualizziamo quindi il segnale su ogni canale e sulla loro somma quadratica. 


```python

########## sfondo ############
sigma = 10
dim = 256

imageR = np.random.randn(dim,dim)
imageI = np.random.randn(dim,dim)
imageM = np.sqrt(np.square(imageR)+np.square(imageI))

countsR, binsR = np.histogram(imageR, bins = 30)
countsI, binsI = np.histogram(imageI, bins = 30)
countsM, binsM = np.histogram(imageM, bins = 30)

fig, axes = plt.subplots(1,3)
axes[0].hist(binsR[:-1], binsR, weights=countsR)
axes[1].hist(binsI[:-1], binsI, weights=countsI)
axes[2].hist(binsM[:-1], binsM, weights=countsM)

plt.show()
```

Ora invece, simuliamo il segnale proveniente da un tessuto con S = 100. Simuliamo l'acquisizione su due canali (fase e quadratura) che rappresenteranno la parte reale e la parte immaginaria del segnale acquisito. Visualizziamo quindi il segnale su ogni canale e sulla loro somma quadratica. 

```python
########## immagine ############
sigma = 10
dim = 256

imageR = 100 + np.random.randn(dim,dim)
imageI = 100 + np.random.randn(dim,dim)
imageM = 100 + np.sqrt(np.square(imageR)+np.square(imageI))

countsR, binsR = np.histogram(imageR, bins = 30)
countsI, binsI = np.histogram(imageI, bins = 30)
countsM, binsM = np.histogram(imageM, bins = 30)

fig, axes = plt.subplots(1,3)
axes[0].hist(binsR[:-1], binsR, weights=countsR)
axes[1].hist(binsI[:-1], binsI, weights=countsI)
axes[2].hist(binsM[:-1], binsM, weights=countsM)

plt.show()
```

E' possibile notare dal confronto degli istogrammi che il segnale sullo sfondo non segue una distribuzione Gaussiana, come invece avviene per il segnale proveniente dal tessuto.
