# Effetto della dimensione della ROI sulla stima di M e SD - Notebook 1.5

Import delle librerie necessarie alla simulazione

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.ndimage as ndi
```

```python
# simulated SD
Tsd = 10

# simulated mean
Tm = 300

# plot mean/SD dispersion vs ROI dimension
N = 100

figure, ax = plt.subplots(1,2, constrained_layout=True)
ax[0].set_xlabel('ROI dim')
ax[0].set_ylabel('Mean (imposed 300)')
ax[0].set_ylim(250,350)
ax[1].set_xlabel('ROI dim')
ax[1].set_ylabel('sd (imposed 10)')
ax[1].set_ylim(0,20)

for k in range(N):
    dim = np.array([],int)
    M   = np.array([])
    sd  = np.array([])
    for i in range(1,10):
        # ROI area
        new_el = (i)*(i)
        dim = np.append(dim, new_el) 

        # generate image with gaussian noise
        image = Tm + Tsd*np.random.randn(dim[i-1])
        M  = np.append(M, np.mean(image, None))
        sd = np.append(sd, np.std(image, None))
    ax[0].plot(dim,M)
    ax[1].plot(dim,sd)
plt.show()

```
